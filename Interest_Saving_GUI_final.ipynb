```python
##### from tkinter import *
from tkinter import ttk
from tkinter import messagebox
from tkinter.ttk import Notebook
from tkcalendar import DateEntry
from tkinter.tix import *
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import pandas as pd
import datetime 
import numpy as np
from calendar import isleap


def clear():
    for i in TVExpense.get_children():
        TVExpense.delete(i)
            


def calculate():
    
    principle = float(Principle.get())
    monthly_contribution = float(MonthlySave.get())
    rate = float(Rate.get())
    end_date = int(DateEnd.get())
    start_date = datetime.datetime.strptime(EDate.get(), '%m/%d/%y')
    compound_frq = Compound_Frq.get()

      
    year = []
    deposits = []
    interest = []    
    accrued_interest =[]
    total_balance = []
    total_deposits = []
    
    if monthly_contribution >0:
        for  i in range(0,end_date+1):
            year.append(i)


            if compound_frq == 'Annually':
                if i > 1:
                    deposits.append(monthly_contribution)
                    total_deposits.append(principle + monthly_contribution*(i))
                    ci = ((total_balance[i-1]* (pow((1+rate/100), 1)))  + deposits[i]) 
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i]) 
                elif i == 1:
                    deposits.append(monthly_contribution)
                    total_deposits.append(principle + monthly_contribution*(i))
                    ci = ((principle * (pow((1+rate/100), 1))) + deposits[i])
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i]) 
                else: 
                    deposits.append(principle)
                    total_deposits.append(principle)
                    ci =  (principle * (pow((1+rate/100), i)))
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i-1]) 
                    
                    
            elif compound_frq == 'Semiannually':
                if i > 1:
                    deposits.append(monthly_contribution)
                    total_deposits.append(principle + monthly_contribution*(i))
                    ci = ((total_balance[i-1]* (pow((1+(rate/100)/2), 1*2)))  + deposits[i])
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i]) 
                elif i == 1:
                    deposits.append(monthly_contribution)
                    total_deposits.append(principle + monthly_contribution*(i))
                    ci = ((principle * (pow((1+(rate/100)/2), 1*2))) + deposits[i])
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i]) 
                else: 
                    deposits.append(principle)
                    total_deposits.append(principle)
                    ci =  (principle * (pow((1+(rate/100)/2), i*2)))
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i-1]) 
                    
                    
            elif compound_frq == 'Quarterly':
                if i > 1:
                    deposits.append(monthly_contribution)
                    total_deposits.append(principle + monthly_contribution*(i))
                    ci = ((total_balance[i-1]* (pow((1+(rate/100)/4), 1*4)))  + deposits[i])
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i]) 
                elif i == 1:
                    deposits.append(monthly_contribution)
                    total_deposits.append(principle + monthly_contribution*(i))
                    ci = ((principle * (pow((1+(rate/100)/4), 1*4))) + deposits[i])
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i]) 
                else: 
                    deposits.append(principle)
                    total_deposits.append(principle)
                    ci =  (principle * (pow((1+(rate/100)/4), i*4)))
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i-1]) 
                
                
            elif compound_frq == 'Monthly':
                if i > 1:
                    deposits.append(monthly_contribution)
                    total_deposits.append(principle + monthly_contribution*(i))
                    ci = ((total_balance[i-1]* (pow((1+(rate/100)/12), 1*12)))  + deposits[i])
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i]) 
                elif i == 1:
                    deposits.append(monthly_contribution)
                    total_deposits.append(principle + monthly_contribution*(i))
                    ci = ((principle * (pow((1+(rate/100)/12), 1*12))) + deposits[i])
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i]) 
                else: 
                    deposits.append(principle)
                    total_deposits.append(principle)
                    ci =  (principle * (pow((1+(rate/100)/12), i*12)))
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i-1]) 
            
            
            elif compound_frq == 'Daily':
                if i > 1:
                    deposits.append(monthly_contribution)
                    total_deposits.append(principle + monthly_contribution*(i))
                    ci = ((total_balance[i-1]* (pow((1+(rate/100)/365), 1*365)))  + deposits[i])
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i]) 
                elif i == 1:
                    deposits.append(monthly_contribution)
                    total_deposits.append(principle + monthly_contribution*(i))
                    ci = ((principle * (pow((1+(rate/100)/365), 1*365))) + deposits[i])
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i]) 
                else: 
                    deposits.append(principle)
                    total_deposits.append(principle)
                    ci =  (principle * (pow((1+(rate/100)/365), i*365)))
                    total_balance.append(ci)
                    accrued_interest.append(total_balance[i] - total_deposits[i-1]) 
                    
            
            if i > 0:
                interest.append(accrued_interest[i] - accrued_interest[i-1])         
            else:
                interest.append(accrued_interest[i])

    
    if monthly_contribution == 0:
        for  i in range(0,end_date+1):
            year.append(i)
            deposits.append(monthly_contribution)
            total_deposits.append(monthly_contribution*(i+1))

            if compound_frq == 'Annually':
                if i > 0:
                    ci = total_balance[i-1] * (pow((1+rate/100), 1))
                    total_balance.append(ci)
                else: 
                    ci =  (principle * (pow((1+rate/100), i)))
                    total_balance.append(ci)
            elif compound_frq == 'Semiannually':
                if i > 0:
                    ci = total_balance[i-1]  * (pow((1+(rate/100)/2), 2*1))
                    total_balance.append(ci)
                else:
                    ci = principle * (pow((1+(rate/100)/2), 2*i))
                    total_balance.append(ci)
            elif compound_frq == 'Quarterly':
                if i > 0:
                    ci = total_balance[i-1] * (pow((1+(rate/100)/4), 4*1))
                    total_balance.append(ci)
                else:
                    ci = principle * (pow((1+(rate/100)/4), 4*i))
                    total_balance.append(ci)
            elif compound_frq == 'Monthly':
                if i > 0:
                    ci = total_balance[i-1] * (pow((1+(rate/100)/12), 12*1))
                    total_balance.append(ci)
                else:
                    ci = principle * (pow((1+(rate/100)/12), 12*i))
                    total_balance.append(ci)
            elif compound_frq == 'Daily':
                if i > 0:
                    ci = total_balance[i-1] * (pow((1+(rate/100)/365), 365*1))
                    total_balance.append(ci)
                else:
                    ci = principle * (pow((1+(rate/100)/365), 365*i))
                    total_balance.append(ci)        
            accrued_interest.append(total_balance[i] - principle - total_deposits[i])             
            if i > 0:
                interest.append(accrued_interest[i] - accrued_interest[i-1])         
            else:
                interest.append(accrued_interest[i])


    Compound.set(total_balance[end_date])
    
    data = {'Year': year, 'Deposits': deposits, 'Interest' : interest, 'Total Deposits': total_deposits,
            'Accrued Interest': accrued_interest, 'Total Balance': total_balance}

    df = pd.DataFrame(data)    
    df['Year'] = df['Year'].astype(int)
    df['Deposits'] = df['Deposits'].round(2)
    df['Interest'] = df['Interest'].round(2)
    df['Total Deposits'] = df['Total Deposits'].round(2)
    df['Accrued Interest'] = df['Accrued Interest'].round(2)
    df['Total Balance'] = df['Total Balance'].round(2)
    
    df['Deposits'] = df.apply(lambda x: "{:,}".format(x['Deposits']), axis=1)
    df['Interest'] = df.apply(lambda x: "{:,}".format(x['Interest']), axis=1)
    df['Total Deposits'] = df.apply(lambda x: "{:,}".format(x['Total Deposits']), axis=1)
    df['Accrued Interest'] = df.apply(lambda x: "{:,}".format(x['Accrued Interest']), axis=1)
    df['Total Balance'] = df.apply(lambda x: "{:,}".format(x['Total Balance']), axis=1)
    

    
    for _ in range(len(df.index.values)): # use for loop to get values in each line, _ is the number of line.
        TVExpense.insert('','end',value=tuple(df.iloc[_,[0,1,2,3,4,5]].values)) # [_,[1,2]] represents that you will get the values of second column and third column for each line.
    
    
    
    figure2 = plt.Figure(figsize=(13.4, 5), dpi=80)
    ax2 = figure2.add_subplot(111)
    bar2 = FigureCanvasTkAgg(figure2,bottom_frame)
    bar2.get_tk_widget().grid(row=0, column=0, padx=1, pady=0)
    df2 =  df[['Year', 'Total Balance']]
    df2['Total Balance'] = [float(str(i).replace(",","")) for i in df['Total Balance']]
    df3 = df2[['Year', 'Total Balance']].groupby('Year').sum()
    df3 .plot(kind='bar',color='Brown', legend=True, ax=ax2)
    ax2.set_title('Yearly Balance')
    ax2.set_facecolor('oldlace')
    figure2.tight_layout()
    

            
    
    
main = Tk()
main.title('Interest Calculator')
main.config(bg="navajowhite3")
main.geometry('950x600')
main.state('zoomed')



#### split frame

left_frame = Frame(main, width = 150, height = 505, bg='oldlace')
left_frame.grid(row=0, column=0, padx=5, pady=5)

right_frame = Frame(main, width= 300, height = 500, bg='oldlace')
right_frame.grid(row=0, column=1, padx=5, pady=10)


bottom_frame = Frame(main, width= 900, height = 250, bg='oldlace')
bottom_frame.grid(row=1, column=1, padx=0, pady=10,)



#create labels
Label(left_frame, text="Track Your Savings", font=('Franklin Gothic Book',15, 'bold', 'underline'),background='oldlace', foreground='mistyrose4').grid(row=0, column=0, padx=0, pady=5)
                                          
Label(right_frame, text="Summary Table", font=('Franklin Gothic Book',15, 'bold', 'underline'),background='oldlace', foreground='mistyrose4').grid(row=0, column=1, padx=400, pady=5)
                                                                               
#---------------1-----------------------
LPrinciple = ttk.Label (left_frame, text='Principal Amount', font=('Franklin Gothic Book',15), background='oldlace', foreground = 'mistyrose4')
LPrinciple.grid(row=1, column=0, padx=5, pady=5, sticky='w')

Principle = DoubleVar(master=left_frame)
EPrinciple = Entry(left_frame,  textvariable=Principle , width=25, background = 'white', foreground = 'black',font=('Franklin Gothic Book',12))
EPrinciple.grid(row=1, column=1, padx=5, pady=5, sticky='w')


#---------------2-----------------------
LRate = ttk.Label (left_frame, text='Interest Rate (%)', font=('Franklin Gothic Book',15), background='oldlace', foreground = 'mistyrose4')
LRate.grid(row=2, column=0, padx=5, pady=5, sticky='w')

Rate = DoubleVar(master=left_frame)

##set textvariable if you want to set default value##
ERate = Entry(left_frame,  textvariable=Rate ,  width=25, background = 'white', foreground = 'black',font=('Franklin Gothic Book',12))
ERate.grid(row=2, column=1, padx=5, pady=5, sticky='w')
Rate.set(0.25)


#---------------3-----------------------
LDate = ttk.Label (left_frame, text='Date of Saving', font=('Franklin Gothic Book',15), background='oldlace', foreground = 'mistyrose4')
LDate.grid(row=4, column=0, padx=5, pady=5, sticky='w')

EDate = DateEntry(left_frame,  width=23, background = 'oldlace', foreground = 'mistyrose4',font=('Franklin Gothic Book',12))
EDate.grid(row=4, column=1, padx=5, pady=5, sticky='w')


#--------------4------------------------
LDateEnd = ttk.Label (left_frame, text='# of Years of Saving', font=('Franklin Gothic Book',15), background='oldlace', foreground = 'mistyrose4')
LDateEnd.grid(row=5, column=0, padx=5, pady=5, sticky='w')

DateEnd = DoubleVar(left_frame)

EDateEnd = Entry(left_frame, textvariable=DateEnd,  width=25, background = 'white', foreground = 'black',font=('Franklin Gothic Book',12))
EDateEnd.grid(row=5, column=1, padx=5, pady=5, sticky='w')



#--------------5-------------------------    
LMonthlySave = ttk.Label (left_frame, text='Expected Yearly Saving', font=('Franklin Gothic Book',15), background='oldlace', foreground = 'mistyrose4')
LMonthlySave.grid(row=3, column=0, padx=5, pady=5, sticky='w')

MonthlySave = DoubleVar(left_frame)

EMonthlySave = Entry(left_frame, textvariable=MonthlySave,  width=25, background = 'white', foreground = 'black',font=('Franklin Gothic Book',12))
EMonthlySave.grid(row=3, column=1, padx=5, pady=5, sticky='w')                 
                          
#--------------6-------------------------   
                          
OPTIONS = ["Annually", "Semiannually", "Quarterly",
           "Monthly", "Daily"]


                          
LCompound_Frq = ttk.Label (left_frame, text='Compound Frequency', font=('Franklin Gothic Book',15), background='oldlace', foreground = 'mistyrose4')
LCompound_Frq.grid(row=6, column=0, padx=5, pady=5, sticky='w')

Compound_Frq = StringVar(left_frame)

Compound_Frq.set("Annually") # default value
                          
ECompound_Frq = ttk.Combobox(left_frame, textvariable=Compound_Frq, values=OPTIONS, width=23, background = 'white', foreground = 'black',font=('Franklin Gothic Book',12))
ECompound_Frq.grid(row=6, column=1, padx=5, pady=5, sticky='w')


                          
                          
                          
LCompound = ttk.Label (left_frame, text='Total Balance', font=('Franklin Gothic Book',15), background='oldlace', foreground = 'mistyrose4')
LCompound.grid(row=8, column=0, padx=5, pady=5, sticky='w')

Compound = DoubleVar(master=left_frame)

##set textvariable if you want to set default value##
ECompound = Entry(left_frame,  textvariable=Compound ,  width=25, background = 'white', foreground = 'black',font=('Franklin Gothic Book',12))
ECompound.grid(row=8, column=1, padx=5, pady=5, sticky='w')




# Create style Object
''' Button 1'''
btn1 = Button(left_frame, text = 'Calculate', bg='oldlace', fg='mistyrose4', font =
               ('Franklin Gothic Book', 12, 'bold'), command = calculate)
btn1.grid(row=7, column=1, padx = 5, pady=5, sticky ='w', ipadx=10, ipady=5)


''' Button 2'''
btn2 = Button(right_frame, text = ' Clear ', bg='oldlace', fg='mistyrose4', font =
               ('Franklin Gothic Book', 12, 'bold'), command = clear)
btn2.grid(row=0, column=2, padx = 5, pady=5, sticky ='e', ipadx=10, ipady=5)



TVList = ['Year', 'Deposits', 'Interest', 'Total Deposits', 'Accrued Interest', 'Total Balance']
TVExpense = ttk.Treeview(right_frame, column=TVList, show='headings', height=15)
for i in TVList:
    TVExpense.heading(i, text= i.title())
TVExpense.grid(row=1, column=0, padx = 10, pady=5, sticky ='w', columnspan = 4)
TVExpense.column("Year", width=175, anchor="c")
TVExpense.column("Deposits", width=175, anchor="c")
TVExpense.column("Interest", width=175, anchor="c")
TVExpense.column("Total Deposits", width=175, anchor="c")
TVExpense.column("Accrued Interest", width=175, anchor="c")
TVExpense.column("Total Balance", width=175, anchor="c")






main.mainloop()
    
```
